```python
import subprocess # Import the subprocess module
import sys # Import the sys module

# Auto-install dependencies if missing (for Colab)
try:
    import gensim
except ImportError:
    subprocess.check_call([sys.executable, "-m", "pip", "install", "gensim", "nltk"])

import gensim.downloader as api
from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import STOPWORDS
from gensim.corpora import Dictionary
from gensim.models import TfidfModel
from nltk.stem import WordNetLemmatizer
import nltk; nltk.download('wordnet', quiet=True)

text = ["Gensim is an open-source library for", "unsupervised topic modeling and", "natural language processing."]
lemmatizer = WordNetLemmatizer()

# Tokenize -> Remove Stopwords -> Lemmatize
docs = [[lemmatizer.lemmatize(t) for t in simple_preprocess(d, deacc=True) if t not in STOPWORDS] for d in text]

# Dictionary & BoW
dct = Dictionary(docs)
bow = [dct.doc2bow(d) for d in docs]

# TF-IDF
print("TF-IDF:", *TfidfModel(bow)[bow], sep="\n")
```
